#### Conceptos de POO (Programacion orientada de objetos) en Python

# Lenguaje Python, crear la clase "animal" con los atributos "nombre", "raza" crear el constructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definimos la clase (el molde)
class Animal:
    # El constructor: se encarga de dar valores iniciales a cada objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función (método) para que el animal ladre
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau!💖")

    # Función (método) para que el animal corra
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente.💥")

    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina 💫")

# --- Creación de los objetos (instancias) ---

# Creamos el objeto 'pitbull' con datos específicos
pitbull = Animal("Rocco", "Pitbull")

# Creamos el objeto 'chihuahua' con datos específicos
chihuahua = Animal("Chispita", "Chihuahua")

# creamos el objeto perro callejero con datos
perrocallejero = Animal("Pelusa", "chafa")

# --- Uso de los objetos ---

print(f"Datos del primer perro: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()
pitbull.morder()
print("-" * 20)

print(f"Datos del segundo perro: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()
chihuahua.morder()
print("-" * 20)

print(f"Datos del tercer perro: {perrocallejero.nombre}, Raza: {perrocallejero.raza}")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.morder()
print("programa creado por Emily Barraza NL = 7")
```
